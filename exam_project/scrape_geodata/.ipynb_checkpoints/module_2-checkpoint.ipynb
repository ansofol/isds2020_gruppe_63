```python
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
import requests
import copy
```

```python
url = 'https://www.kmdvalg.dk/fv/2019/KMDValgFV.html'
html = requests.get(url).text
soup = BeautifulSoup(html, 'html')

out_list = list()
item_list = soup.find_all('a', {'class':'list-group-item valg-status-final'})
out_list.append([item_list[item]['href'] for item in range(len(item_list))])
```

```python
def get_refs(url_list):
    out_list = list()
    if isinstance(url_list, str):
        url_list = [url_list]
    for url in url_list:
        html = requests.get(url).text
        soup = BeautifulSoup(html, 'html')
        
        #Tjekker om links refererer til kommuner eller valgsteder:
        check_link_cat = soup.find_all('div', {'class':'content-block-header'}) 
        if (True in ["Kommuner" in check_link_cat[i].text for i in range(len(check_link_cat))]):
            link_cat = "Kommune"
        elif (True in ["Storkredse og valgkredse" in check_link_cat[i].text for i in range(len(check_link_cat))]):
            link_cat = "Valgkreds"
        elif (True in ["Afstemningssteder" in check_link_cat[i].text for i in range(len(check_link_cat))]):
            link_cat = "Valgsted"
        else:
            link_cat = "Fejl"
        
        item_list = soup.find_all('a', {'class':'list-group-item valg-status-final'})
        link_pair = [[link_cat, item_list[item]['href']] for item in range(len(item_list))]
        out_list.extend(link_pair)
        
    out_df = pd.DataFrame(out_list)
    return out_df

def construct_links(ref_list, baselink):
    if isinstance(ref_list, str):
        out_list = baselink+ref_list
    else:
        out_list = [baselink+ref_list[item] for item in range(len(item_list))]
    return out_list
```
